```python
import pandas as pd
import os
import shutil
from pathlib import Path

# Cargar el CSV
csv_path = "/train_fullset_patches/gt.csv"
df = pd.read_csv(csv_path)

# Crear carpeta de destino si no existe
dest_dir = "train_sample"
os.makedirs(dest_dir, exist_ok=True)

# Asumiendo que las imágenes están en esta carpeta
source_dir = "train_fullset_patches"

# Seleccionar 4 imágenes por cada label
for label in range(1, 7):
    subset = df[df['labels'] == label]
    sample_images = subset.sample(n=4, random_state=42)  # Fijamos la semilla para reproducibilidad

    for _, row in sample_images.iterrows():
        image_name = row['images']
        src_path = os.path.join(source_dir, image_name)
        dst_path = os.path.join(dest_dir, image_name)

        # Copiar la imagen si existe
        if os.path.isfile(src_path):
            shutil.copy2(src_path, dst_path)
        else:
            print(f"Imagen no encontrada: {src_path}")
```
